```python
import pandas as pd

mkstudio_df = pd.read_csv('billboard200.csv')
display(mkstudio_df)

```

```python
mkstudio_df = mkstudio_df[['Rank', 'Artist', 'Peak Position']]
display(mkstudio_df)
```


Listar os top 10 artistas que começam com a letra "P", seguindo o seguinte critério de pontuação: Cada música do artista que fica por 1 semana no billboard charts dá 1 ponto para esse artista.
Por exemplo, o artista Prodigy tem 3 músicas na base de dados, H-N-I-C = 15 semanas, Return Of The Mac = 3 semanas,H.N.I.C. Pt. 2 = 2 semanas. Nesse critério Prodigy tem 20 pontos.

```python
nome_p = mkstudio_df["Artist"].str.startswith('P', na=False)
mkstudio_df[nome_p]
```

```python

seg_saida = mkstudio_df[nome_p]
seg_saida = seg_saida.groupby('Artist').sum()
display(seg_saida)
```

```python

seg_saida = seg_saida.sort_values('Peak Position', ascending=False)
display(seg_saida[:10])
```
